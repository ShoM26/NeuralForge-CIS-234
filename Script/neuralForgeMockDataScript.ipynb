# **Creating the mock data:**
Goal: create 2 functions that have loops that run n times; each time generating realistic data for the designated table that adds a string with json formatted values to a collection which will be passed to the api call below.

Example Json format: {"variable_name_1": "value_1", "variable_name_2": "value_2"}

EventType
    PowerLost=1,
    Maintenance=2,
    RoutineCleaning=3

```python
import json
import random
from datetime import datetime, timedelta

def generate_production_records(lines):
    collection = []
    current_time = datetime.now().replace(minute=0, second=0, microsecond=0)
    hours = 12
    for i in range(hours):
      for m in range(1, lines+1):
          data = {
              "assemblyLineId": m,
              "numberOfChips": random.randint(35, 60),
              "hourOfDay": (current_time + timedelta(hours=i)).isoformat()
          }
          collection.append(data)

    return collection

def generate_downtime_events(lines):
    collection = []
    for i in range(1, lines+1):
        if random.random() < .75: # 25 percent chance there is an issue
            continue
        event_type = random.randint(1,3)
        start_time = datetime.now().replace(hour=random.randint (1,23))
        end_time = start_time + timedelta(hours = random.randint(2, 5))
        data = {
            "assemblyLineId": i,
            "startTime": start_time.isoformat(),
            "endTime": end_time.isoformat(),
            "eventType": event_type
            }
        collection.append(data)

    return collection
```

# **Calling the c\# api endpoint:**
Goal: take the collection of strings above and using the python json library convert each string to json then call the api endpoint and wait 3 seconds before the next loop iteration

Example Json: The function you're looking for is json.loads(<string_variable>) which returns the raw json data

```python
import json
import requests
from time import sleep

def post_request_production_records(collection):
  endpoint = "http://localhost:5115/api/ProductionRecord"
  for item in collection:
    try:
      response = requests.post(
          endpoint,
          json=item
      )

      if response.status_code in [200,201]:
        print(f"Success (Status: {response.status_code})")
      else:
        print(f"Failed (Status: {response.status_code})")
      sleep(1)
    except Exception as e:
      print(f"Error: {str(e)}")

def post_request_downtime_events(collection):
  endpoint = "http://localhost:5115/api/DowntimeEvent"
  for item in collection:
    try:
      response = requests.post(
          endpoint,
          json=item
      )

      if response.status_code in [200,201]:
        print(f"Success (Status: {response.status_code})")
      else:
        print(f"Failed (Status: {response.status_code})")
      sleep(1)
    except Exception as e:
      print(f"Error: {str(e)}")
```

```python
#Main method that calls the string creation and api requests
if __name__ == "__main__":
  number_of_lines = 70
  production_record_strings = generate_production_records(number_of_lines)
  downtime_event_strings = generate_downtime_events(number_of_lines)
  post_request_production_records(production_record_strings)
  post_request_downtime_events(downtime_event_strings)
```
